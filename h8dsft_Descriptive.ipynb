```python
import pandas as pd
import numpy as np
```

```python
#data set
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')

df.tail(50)
```

## Problem
1. Cari Measure of Central Tendency dari data tersebut
2. Cari Measure of Variance dari data tersebut

```python
df.info()
```

## Problem 1 : Mencari Measure of Central Tendency

Mencari Mean dari data

```python
#assign kolom yang diinginkan ke 1 variabel
kolor = ['New Cases', 'New Deaths', 'New Recovered', 'New Active Cases', 'Total Cases', 'Total Deaths', 'Total Recovered', 'Total Active Cases']
```

```python
for i in kolor:
    print(df[i].name,':',df[i].mean())
```

Mencari median

```python
for i in kolor:
    print(df[i].name,':',df[i].median())
```

Mencari mode

```python
for i in kolor:
    print(df[i].name,':',df[i].mode()[0])
```

## Problem 2 : Mencari Measure of Variance

```python
#melihat konten data set kembali
df.info()
```

Mencari range dari kolom dengan nilai int

```python
#loop display
for i in kolor:
    print(df[i].name,'range: ',(df[i].max()-df[i].min()))
```

Mencari quartil bawah, tengah dan atas

```python
for i in kolor:
    print(df[i].name,':\n',df[i].quantile([0.25, 0.5, 0.75]))
    print('\n')
```

Mencari variance

```python
for i in kolor:
    print(df[i].name,': ',df[i].var())
```

Mencari standard deviataion

```python
for i in kolor:
    print(df[i].name, ':', df[i].std())
```
