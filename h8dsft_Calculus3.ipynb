## Non Graded Challenge 8 - Partial Derivative

Cari gradient dari fungsi berikut:

$$3x+4y=5$$

persamaan menjadi :$$y=-0.75x+1.25$$

```python
#import lib
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define x as variable symbol
x = sy.Symbol('x', real=True)

#y function
y = -0.75*x+1.25
```

```python
def plot(func):
    x = np.linspace(-5,5,50)
    y = func(x)
    plt.plot(x, y, '-.c', label='y=-0.75+1.25')
    plt.title('Graph of y=-0.75x+1.25')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show

plot(lambda x: -0.75*x+1.25)
```

Karena persamaan adalah persamaan linier, maka nilai gradien dapat diperoleh dengan menurunkan persamaan sekali

```python
gradien = sy.diff(y)
gradien
```
